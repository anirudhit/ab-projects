```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.target_names
```

```python
print(digits.target)
print(len(digits.target))
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df.shape
```

```python
df['target'] = digits.target
df.head()
```

```python
digits.target_names
```

```python
df[df.target==3].head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.head()
```

```python
X = df.drop(['target'], axis='columns')
X.head()
```

```python
y = df.target
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
y_train.head()
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model_c = SVC(C=1,kernel='linear')
model_c.fit(X_train, y_train)
```

```python
model_c.score(X_test, y_test)
```
